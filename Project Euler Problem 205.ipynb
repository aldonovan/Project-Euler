```python
import random
def simulateGame():
    sum1 = 0
    for i in range(9):
        sum1 += random.randint(1, 4)
    sum2 = 0
    for j in range(6):
        sum2 += random.randint(1, 6)
    if(sum1 > sum2):
        return True
    return False

numWins = 0
for x in range(1000000):
    if(simulateGame()):
        numWins += 1
    if(x % 10000 == 0):
        print(x)
print(numWins)
```
